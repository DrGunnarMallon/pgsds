```python
from openai import OpenAI
client = OpenAI()

topic="folium"
course = "Python for Geospatial Data Science and Visualization"

completion = client.chat.completions.create(
  model="gpt-3.5-turbo",
  messages=[
    {"role": "system", "content": "You are a content creator, python expert and computer science instructor with decades of experience producing high quality python programming tutorials."},
    {"role": "user", "content": f"Your task is to produce the outline and structure for a lecture on the {topic}. It should be very detailed with many examples and exercies. The lecture is part of a larger course called {course}. The answer should be in Markdown format."}
  ]
)

print(completion.choices[0].message.content)
```
